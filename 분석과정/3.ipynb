```python
df['전북'].plot(figsize=(12,5))
```

```python
from prophet import Prophet

df['ds']= df['기준일']
df['y'] = df['전북']


model = Prophet()

model.fit(df)

# 2023년 4월부터 2026년 3월까지 36개월 동안 예측
future = model.make_future_dataframe(periods=33, freq='M')

# 예측 수행
forecast = model.predict(future)


# 결과 출력
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']])

forecast[['yhat','yhat_lower','yhat_upper']].plot(figsize=(12,5))
```

```python
# 예측 결과를 Excel 파일로 저장
forecast.to_excel('향후 3년간 전기차 대수 예측 결과.xlsx', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 데이터를 DataFrame으로 생성
data = {
    '기준일': [
        '2022-01-31', '2022-02-28', '2022-03-31', '2022-04-30', '2022-05-31',
        '2022-06-30', '2022-07-31', '2022-08-31', '2022-09-30', '2022-10-31',
        '2022-11-30', '2022-12-31', '2023-01-31', '2023-02-28', '2023-03-31'
    ],
    '전북': [7389, 7456, 8404, 8931, 9310, 9723, 10246, 11353, 12162, 12448, 12622, 12727, 12740, 13022, 15070]
}

df = pd.DataFrame(data)

# 기준일 컬럼을 datetime 형식으로 변환
df['기준일'] = pd.to_datetime(df['기준일'])

# 기준일을 인덱스로 설정
df.set_index('기준일', inplace=True)

# ARIMA 모델 피팅
model = ARIMA(df['전북'], order=(1, 1, 1))  # (p, d, q) 파라미터 설정
model_fit = model.fit()

# 2023년 4월부터 2027년 1월까지 46개월 동안 예측
forecast = model_fit.forecast(steps=46)

# 예측 결과 출력
print(forecast)

# 예측 결과의 길이를 실제 값과 맞춤

actual_values = df['전북'].values[-len(forecast):]
forecast = forecast[:len(actual_values)]

mse = mean_squared_error(actual_values, forecast)
mae = mean_absolute_error(actual_values, forecast)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")

# charging 데이터
charging_dates = ['2024-01-01', '2025-01-01', '2026-01-01', '2027-01-01']
charging_values = [2730.46672222*5, 3280.50415079*5, 3690.64357937*5, 3920.54366667*5]



# 예측 결과를 시각화
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['전북'], label='Actual', color='#A7E097')
plt.plot(pd.date_range(start='2023-04-30', periods=46, freq='M'), forecast, label='Forecast', color='#43942C')
plt.bar(pd.to_datetime(charging_dates), charging_values,width=50, color='#C73C43', label='Charging')
plt.xlabel('Date')
plt.ylabel('Number of EVs')
plt.legend()
plt.title('EV Forecast using ARIMA')
plt.show()
```
